## Summary

- **Step 1:** Imported the dataset using `pandas`.
- **Step 2:** Cleaned the `discounted_price` and `actual_price` columns by removing the ₹ symbol and commas and converting them to floats.
- **Step 3:** Processed the `discount_percentage` column by removing the "%" symbol and converting to a decimal.
- **Step 4:** Converted the `rating` column to numeric, coercing any non-numeric entries to NaN.
- **Step 5:** Cleaned the `rating_count` column by removing commas, converting to integer, and filling missing values.
- **Step 6:** Computed and printed basic statistics for the numerical columns.
- **Step 7:** Saved the cleaned dataset to a new CSV file for further analysis.



```python
# Clean and convert columns by first converting to string
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('[₹,]', '', regex=True).astype(float)
df['actual_price'] = df['actual_price'].astype(str).str.replace('[₹,]', '', regex=True).astype(float)
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '', regex=False).astype(float) / 100

# Convert 'rating' to numeric; non-numeric values will become NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# For 'rating_count', first convert to string then remove commas and convert to numeric
df['rating_count'] = pd.to_numeric(df['rating_count'].astype(str).str.replace(',', '', regex=False), errors='coerce')

# Handle missing values by using assignment rather than inplace to avoid chained assignment warnings
df['rating'] = df['rating'].fillna(df['rating'].median())
df['rating_count'] = df['rating_count'].fillna(0)

# Compute and display basic statistics for the numerical columns
print("Statistics for numerical columns:")
print(df[['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].describe())
```

# Grouping Data by Category

In this section, we perform groupings on the categorical column (`category`). Our objective is to calculate:
- **Mean rating per category**
- **Mean rating_count per category**

Given that the `category` column has 211 unique values, this analysis helps to identify which categories have the best and worst average ratings, and how many ratings they attract.

The following code groups the dataset by `category` and computes the required mean values.


```python
# Group by the 'category' column and calculate the mean for 'rating' and 'rating_count'
category_stats = df.groupby('category')[['rating', 'rating_count']].mean()

# Display the resulting statistics
print("Mean rating and rating count per category:")
print(category_stats)
```

# Identifying Patterns and Interesting Findings

In this section, we investigate potential patterns in the dataset:

- **High/Low Ratings:**  
  Examine whether certain categories consistently receive higher ratings. For instance, do electronics outperform other categories in ratings?

- **Rating Count Trends:**  
  Assess if categories with more ratings (i.e., higher `rating_count`) tend to have higher or lower average ratings.

- **Discount vs. Rating:**  
  Explore if there's a relationship between the `discount_percentage` and the `rating`.  
  To analyze this, we compute the correlation between these two columns.

The correlation value will indicate:
- A value near 0: No linear relationship.
- A positive value: As discounts increase, ratings tend to be higher.
- A negative value: As discounts increase, ratings tend to be lower.

The following Python code computes this correlation.


```python
# Compute the correlation between discount_percentage and rating
correlation = df['discount_percentage'].corr(df['rating'])
print(f"Correlation between discount percentage and rating: {correlation:.2f}")
```

# Data Visualization: Cumulative Rating Count

In this section, we create a line chart that shows the cumulative rating count across products. Although our dataset isn’t time-series data, sorting by `product_id` gives us a logical order to assess how customer engagement (via ratings) accumulates across the product list.

### Steps:
1. **Sort the Data:**  
   The dataframe is sorted based on `product_id` to establish a consistent product order.

2. **Plot the Cumulative Rating Count:**  
   The cumulative sum of the `rating_count` column is plotted against the product index.
   
3. **Insight:**  
   This chart highlights products that receive significant customer feedback by visualizing where the cumulative rating count increases rapidly.




```python
import matplotlib.pyplot as plt

# Sort the DataFrame by product_id to establish order
df_sorted = df.sort_values('product_id')

# Plot the cumulative sum of rating_count
plt.figure(figsize=(10, 6))
plt.plot(df_sorted.index, df_sorted['rating_count'].cumsum(), color='blue')
plt.title('Cumulative Rating Count Across Products')
plt.xlabel('Product Index')
plt.ylabel('Cumulative Rating Count')
plt.grid(True)
plt.show()
```

# Exploratory Data Analysis (EDA)

The objective of this analysis is to uncover insights into pricing, ratings, and customer behavior—key elements in data-driven decision-making.

### Pricing Insights:
- **Average Discount:**  
  Calculate the mean discount percentage.
- **Top Discounted Products:**  
  Identify the top 5 products with the highest discount percentages, including their names.

### Rating Insights:
- **Average Rating:**  
  Calculate the average rating across all products.
- **Distribution of Ratings:**  
  Generate descriptive statistics of the ratings (e.g., count, mean, std, min, max).

### Category Insights:
- **Most Common Categories:**  
  Identify the most frequent categories in the dataset.

The following code demonstrates these analyses.


```python
# Pricing Insights
avg_discount = df['discount_percentage'].mean() * 100  # converting to percentage
top_discounted = df.nlargest(5, 'discount_percentage')[['product_name', 'discount_percentage']]

# Rating Insights
avg_rating = df['rating'].mean()
rating_distribution = df['rating'].describe()

# Category Insights
most_common_categories = df['category'].value_counts().head()

# Print the insights
print(f"Average discount: {avg_discount:.5f}%")
print("\nTop 5 products by discount:")
print(top_discounted)
print(f"\nAverage rating: {avg_rating:.2f}")
print("\nRating distribution:")
print(rating_distribution)
print("\nMost common categories:")
print(most_common_categories)
```

# Data Visualization (Top 10 Items)

The objective is to visualize key trends while only displaying the top 10 items for each chart. This approach helps focus on the most significant data points.

### 1. Discount vs. Rating Scatter Plot (Top 10 by Discount)
- **Objective:**  
  To examine the relationship between discount percentage and rating for the top 10 products with the highest discounts.
- **Method:**  
  We filter the dataset to include only these top 10 products and then plot their discount percentages versus ratings.

### 2. Category Distribution Bar Chart (Top 10 Categories)
- **Objective:**  
  To display the top 10 product categories by count.
- **Method:**  
  We compute the frequency of each category and then plot the top 10 most common ones.

### 3. Rating Distribution (Top 10 Most Frequent Ratings)
- **Objective:**  
  To understand which rating values occur most frequently.
- **Method:**  
  Instead of a continuous histogram, we display a bar chart of the top 10 most frequent ratings.


```python
import matplotlib.pyplot as plt

# 1. Discount vs. Rating Scatter Plot (Top 10 by Discount)
df_top10_discount = df.nlargest(10, 'discount_percentage')
plt.figure(figsize=(8, 6))
plt.scatter(df_top10_discount['discount_percentage'], df_top10_discount['rating'], alpha=0.5, color='teal')
plt.xlabel('Discount Percentage')
plt.ylabel('Rating')
plt.title('Discount vs. Rating (Top 10 Products by Discount)')
plt.grid(True)
plt.show()

# 2. Category Distribution Bar Chart (Top 10 Categories)
plt.figure(figsize=(10, 6))
df['category'].value_counts().head(10).plot(kind='bar', color='coral')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.title('Top 10 Categories by Product Count')
plt.tight_layout()
plt.show()

# 3. Rating Distribution (Top 10 Most Frequent Ratings)
plt.figure(figsize=(8, 6))
# Get the top 10 most frequent rating values and sort by the rating value
df['rating'].value_counts().head(10).sort_index().plot(kind='bar', color='mediumseagreen')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Top 10 Most Frequent Ratings')
plt.tight_layout()
plt.show()
```
